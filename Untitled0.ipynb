```python
!pip install tensorflow tensorflow-datasets matplotlib numpy Pillow tqdm ipython
!pip install PyGithub

from IPython import get_ipython
from IPython.display import display

import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np
import os
import time
from IPython.display import display
from PIL import Image
from tqdm.notebook import tqdm

# Bildauflösung und Batch-Größe
IMG_WIDTH = 128
IMG_HEIGHT = 128
BATCH_SIZE = 8

# Datensatz laden und vorbereiten
dataset, info = tfds.load('oxford_flowers102', with_info=True, as_supervised=True)
train_dataset = dataset['train']

# Bildvorverarbeitungsfunktion
def preprocess(image, label):
    image = tf.image.resize(image, [IMG_WIDTH, IMG_HEIGHT])
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

# Datenaugmentation
def augment_image(image, label):
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_crop(image, size=[IMG_WIDTH, IMG_HEIGHT, 3])
    return image, label

# Vorverarbeitung, Augmentation und Batching anwenden
train_dataset = train_dataset.map(preprocess).map(augment_image).cache().batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# Generator-Modell mit ResNet-Blöcken und angepasster Auflösung
latent_dim = 200

def resnet_block(input_tensor, filters):
    x = tf.keras.layers.Conv2D(filters, (3, 3), padding='same')(input_tensor)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = tf.keras.layers.Conv2D(filters, (3, 3), padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Add()([x, input_tensor])  # Skip connection
    x = tf.keras.layers.ReLU()(x)
    return x

def make_generator_model():
    inputs = tf.keras.layers.Input(shape=(latent_dim,))

    initial_size = min(IMG_WIDTH, IMG_HEIGHT) // 16
    x = tf.keras.layers.Dense(initial_size * initial_size * 512)(inputs)

    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = tf.keras.layers.Reshape((initial_size, initial_size, 512))(x)

    num_upsamples = int(np.log2(max(IMG_WIDTH, IMG_HEIGHT) / initial_size))

    filters = 512
    for i in range(num_upsamples):
        x = tf.keras.layers.Conv2DTranspose(filters, (5, 5), strides=(2, 2), padding='same')(x)
        x = tf.keras.layers.BatchNormalization()(x)
        x = tf.keras.layers.ReLU()(x)
        x = resnet_block(x, filters)
        x = resnet_block(x, filters)

        x = tf.keras.layers.Conv2D(filters, (3, 3), padding='same')(x)
        x = tf.keras.layers.BatchNormalization()(x)
        x = tf.keras.layers.ReLU()(x)

        filters = filters // 2 if filters > 32 else filters

    outputs = tf.keras.layers.Conv2DTranspose(3, (5, 5), strides=(1, 1), padding='same', activation='sigmoid')(x)

    model = tf.keras.Model(inputs=inputs, outputs=outputs)
    return model

generator = make_generator_model()

# Diskriminator-Modell
discriminator = tf.keras.Sequential([
    tf.keras.layers.Input(shape=[IMG_WIDTH, IMG_HEIGHT, 3]),
    tf.keras.layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.LeakyReLU(),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Conv2D(64, (3, 3), padding='same'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.LeakyReLU(),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.LeakyReLU(),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Conv2D(128, (3, 3), padding='same'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.LeakyReLU(),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Conv2D(256, (5, 5), strides=(2, 2), padding='same'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.LeakyReLU(),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Conv2D(256, (3, 3), padding='same'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.LeakyReLU(),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(1)
])

# Gradient Penalty
def gradient_penalty(real_img, fake_img):
    epsilon = tf.random.uniform([BATCH_SIZE, 1, 1, 1], 0.0, 1.0)
    interpolated_img = epsilon * real_img + (1 - epsilon) * fake_img

    with tf.GradientTape() as gp_tape:
        gp_tape.watch(interpolated_img)
        pred = discriminator(interpolated_img, training=True)

    grads = gp_tape.gradient(pred, [interpolated_img])[0]
    norm = tf.sqrt(tf.reduce_sum(tf.square(grads), axis=[1, 2, 3]))
    gp = tf.reduce_mean((norm - 1.0) ** 2)
    return gp

# Loss-Funktionen
def discriminator_loss(real_img, fake_img):
    real_loss = tf.reduce_mean(discriminator(real_img))
    fake_loss = tf.reduce_mean(discriminator(fake_img))
    return fake_loss - real_loss

def generator_loss(fake_img):
    return -tf.reduce_mean(discriminator(fake_img))

# Lernraten-Scheduler
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=1e-4,
    decay_steps=10000,
    decay_rate=0.96)

generator_optimizer = tf.keras.optimizers.Adam(lr_schedule)
discriminator_optimizer = tf.keras.optimizers.Adam(lr_schedule)

# Verzeichnisse
save_model_dir = './saved_models'
os.makedirs(save_model_dir, exist_ok=True)
checkpoint_dir = './training_checkpoints'
os.makedirs(checkpoint_dir, exist_ok=True)

# Trainingsparameter
EPOCHS = 200
noise_dim = latent_dim
num_examples_to_generate = 4
seed = tf.random.normal([num_examples_to_generate, noise_dim])

# Funktion zum Generieren und Plotten von Bildern
def generate_and_plot_images(model, epoch, test_input, save_dir, num_rows=4, num_cols=4, border_width=5, scale_factor=2):
    predictions = model(test_input, training=False)
    fig = plt.figure(figsize=(num_cols * scale_factor, num_rows * scale_factor))
    for i in range(predictions.shape[0]):
        plt.subplot(num_rows, num_cols, i + 1)
        plt.imshow(predictions[i, :, :, :])
        plt.axis('off')

    plt.tight_layout(pad=border_width)
    plt.savefig(os.path.join(save_dir, f'epoch_{epoch}_images.png'))
    plt.show()
    plt.close(fig)

# **Änderung hier:** Force optimizer variable creation before training loop
_ = generator(tf.random.normal([1, noise_dim]))  # Force optimizer variable creation
_ = discriminator(tf.random.normal([1, IMG_WIDTH, IMG_HEIGHT, 3]))  # Force optimizer variable creation


# Trainingsschleife - ohne @tf.function
def train_step(images, train_generator=True):
    noise = tf.random.normal([BATCH_SIZE, noise_dim])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        generated_images = generator(noise, training=True)

        real_output = discriminator(images, training=True)
        fake_output = discriminator(generated_images, training=True)

        gen_loss = generator_loss(fake_img=generated_images)
        disc_loss = discriminator_loss(real_img=images, fake_img=generated_images)

        # Gradient Penalty
        gp = gradient_penalty(images, generated_images)
        disc_loss += 10 * gp

    if train_generator:
        gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
        # Gradienten anwenden außerhalb von tf.function
        generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    else:
        gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)
        # Gradienten anwenden außerhalb von tf.function
        discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))

    return gen_loss, disc_loss

def train(dataset, epochs):
    # TensorBoard Integration
    writer = tf.summary.create_file_writer('./logs')
    tf.summary.trace_on(graph=True)

    for epoch in range(epochs):
        start = time.time()

        epoch_gen_losses = []
        epoch_disc_losses = []

        with tqdm(total=tf.data.experimental.cardinality(train_dataset).numpy(), desc=f"Epoch {epoch + 1}/{EPOCHS}") as pbar:
            for image_batch, _ in train_dataset:
                # Separate Training Loops
                for _ in range(5):
                    gen_loss, disc_loss = train_step(image_batch, train_generator=False)
                gen_loss, disc_loss = train_step(image_batch, train_generator=True)

                epoch_gen_losses.append(gen_loss.numpy())
                epoch_disc_losses.append(disc_loss.numpy())
                pbar.update(1)

                # TensorBoard logging
                with writer.as_default():
                    tf.summary.scalar('generator_loss', gen_loss, step=epoch)
                    tf.summary.scalar('discriminator_loss', disc_loss, step=epoch)

        # Modell speichern alle 15 Epochen
        if (epoch + 1) % 15 == 0:
            checkpoint.save(file_prefix = checkpoint_prefix)

        avg_gen_loss = np.mean(epoch_gen_losses)
        avg_disc_loss = np.mean(epoch_disc_losses)

        print ('Time for epoch {} is {} sec - Generator loss: {} - Discriminator loss: {}'.format(epoch + 1, time.time()-start, avg_gen_loss, avg_disc_loss))

        generate_and_plot_images(generator,
                                 epoch + 1,
                                 seed,
                                 checkpoint_dir)

# Checkpoint
checkpoint_prefix = os.path.join(checkpoint_dir, "ckpt")
checkpoint = tf.train.Checkpoint(generator_optimizer=generator_optimizer,
                                 discriminator_optimizer=discriminator_optimizer,
                                 generator=generator,
                                 discriminator=discriminator)

train(train_dataset, EPOCHS)

# Finale Modelle speichern
final_generator_path = os.path.join(save_model_dir, 'final_generator.keras')
final_discriminator_path = os.path.join(save_model_dir, 'final_discriminator.keras')
generator.save(final_generator_path)
discriminator.save(final_discriminator_path)
```
